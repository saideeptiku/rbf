```python
import pandas as pd
df = pd.read_csv('RBF/data/bcinfill-s6-run1.csv') # give the input files here
df2 = pd.read_csv('RBF/data/bcinfill-s6-run2.csv')
```

```python
dfeid = df.loc[df['ssid'] == 'csu-eid']
dfeid2 = df2.loc[df2['ssid'] == 'csu-eid']
```

```python
xcoordinates = dfeid.x.unique()
ycoordinates = dfeid.y.unique()
bssidunique1 = dfeid.bssid.unique()
bssidunique2 = dfeid2.bssid.unique()
dfeid
```

```python
mappairs = {}
APNO = 0
for i in bssidunique1:
    flag = i in mappairs
    if flag == False:
        mappairs[i] = APNO
        APNO = APNO+1
        
for j in bssidunique2:
    flag = j in mappairs
    if flag == False:
        mapppairs[j] = APNO
        APNO = APNO+1
```

```python
macid = list(mappairs.values())
columns = macid + ['x', 'y','timestamp']
leng = len(xcoordinates)
leng
```

```python
leng = leng+1
# the training dataframe, contains only the training data set
newdf = pd.DataFrame( index = range(0,leng), columns = columns)


# the testing datfram, contains only the testing data set
newdftesting = pd.DataFrame( index = range(0,leng), columns = columns)
```

```python
i =0;
previousx =0;
for index,row in dfeid.iterrows():
    x = row['x']
    y = row['y']
    rssi = row['level']
    bssid = row['bssid']
    tst = row['Timestamp']
    value = mappairs[bssid]
    if x == previousx:
        newdf['x'].iloc[i] = x
        newdf['y'].iloc[i] = y
        newdf[value].iloc[i] = rssi
        newdf['timestamp'].iloc[i] = tst
    else:
        i = i+1;
        newdf['x'].iloc[i] = x
        newdf['y'].iloc[i] = y
        newdf[value].iloc[i] = rssi
        newdf['timestamp'].iloc[i] = tst
    previousx = x;
```
